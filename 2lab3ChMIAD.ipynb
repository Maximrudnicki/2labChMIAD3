```python
import pandas as pd
import matplotlib.pyplot as plt
```

12. Сгенерировать или импортировать данные из произвольного источника. Построить 3 разных парных графика, добавить подписи значений и категорий, выставить заголовок, легенду, экспортировать (программным образом) в виде файла изображения.

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(12,5))

plt.plot(gas['Year'], gas['USA'], 'r.-', label='United States')
plt.plot(gas['Year'], gas['Canada'], 'b.-' ,label='Canada')

plt.title('USA vs Canada Gas Prices')
plt.ylabel('US Dollars/Gallon')
plt.legend()

plt.xticks(gas['Year'])

plt.savefig('USA Canada.jpg', dpi=250)

plt.show()
```

```python
def addlabels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')

mx = round((gas['Mexico'].mean()),2)
sk = round((gas['South Korea'].mean()), 2)

x = ['Мексика', 'Южная Корея']
y = [mx, sk]

plt.title('Средняя цена на бензин в Мексике и Южной Корее')
plt.bar(['Мексика', 'Южная Корея'], [mx, sk], label='$')

addlabels(x, y)

plt.ylabel('Цена за галлон бензина')

plt.legend()

plt.savefig('Mexico South Korea.png', dpi=250)

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.scatter(gas['Year'], gas['Germany'], color='lime', label='Germany')
plt.scatter(gas['Year'], gas['South Korea'], color='purple', label='South Korea')

plt.title('Germany vs South Korea Gas Prices')
plt.ylabel('US Dollars/Gallon')
plt.legend()

plt.xticks(gas['Year'])

plt.savefig('Germany South Korea.jpg', dpi=250)

plt.show()
```

13. Сгенерировать или импортировать данные из произвольного источника. Построить 3 разных диаграммы расходящихся столбцов и точек, добавить подписи значений и категорий, выставить заголовок, легенду, экспортировать (программным образом) в виде файла изображения.

```python
india = pd.read_csv('India_GDP_Data.csv')
india
```

```python
colors = [
    'red' if x < 0 else 'green' for x in india['Percentage_Growth '][::2]
]

plt.figure(figsize=(15,25), dpi=100)
plt.hlines(y=india['Year'][::2], xmin=0, xmax=india['Percentage_Growth '][::2],
           colors=colors, label='GDP Growth in %')

plt.title('Growth of India', fontdict={'fontsize': 20})
plt.yticks(india['Year'][::2])
plt.xlabel('Percentage %', fontdict={'fontsize': 20})
plt.ylabel('Year', fontdict={'fontsize': 20})
for x, y, txt in zip(india['Percentage_Growth '][::2],india['Year'][::2],india['Percentage_Growth '][::2]):
    plt.text(x, y, txt, horizontalalignment='right' if x < 0 else 'left',
                 verticalalignment='center', fontdict={'color':'red' if x < 0 else 'green', 'size':14})
plt.legend()

plt.savefig('Growth of India.png', dpi=500)

plt.show()
```

```python
color = [
    'red' if x < 0 else 'green' for x in india['Percentage_Growth ']
]

plt.figure(figsize=(10,25))
plt.scatter(india['Percentage_Growth '], india['Year'],
            s=450, alpha=.9, label='GDP Growth in %', color=color)
for x, y, txt in zip(india['Percentage_Growth '],india['Year'],india['Percentage_Growth ']):
    plt.text(x, y, txt, horizontalalignment='center',
                 verticalalignment='center', fontdict={'color':'w'})

plt.gca().spines["top"].set_alpha(.3)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(.3)
plt.gca().spines["left"].set_alpha(.3)


plt.title('Growth of India', fontdict={'fontsize': 20})
plt.yticks(india['Year'])
plt.xlabel('Percentage %', fontdict={'fontsize': 20})
plt.ylabel('Year', fontdict={'fontsize': 20})

plt.grid(linestyle='--', alpha=0.5)
plt.legend()


plt.savefig('india.jpg')
plt.show()
```
